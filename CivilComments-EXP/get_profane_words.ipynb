### Using LLM converting from toxic to non-toxic fr contrastive learning

````python
from groq import Groq

client = Groq(api_key='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')
````

````python
def generate_non_toxic(toxic_sentence, model="llama3-8b-8192", temperature=0.0):
    prompt = f"""
    Given the following toxic sentence, rewrite it to be non-toxic, maintaining the same meaning or context but with a supportive, neutral, or constructive tone. 
    Ensure the new sentence avoids harsh, condescending, or negative phrasing.Dont provide any additional explanation.
    
    Toxic Sentence: "{toxic_sentence}"
    """
    try:
        response = client.chat.completions.create(
            messages=[
                {"role": "user", "content": prompt}
            ],
            model=model,
            temperature=temperature,
            max_tokens=200
        )
        return response.choices[0].message.content.strip()
    except Exception as e:
        print(f"Error processing: {toxic_sentence}\n{e}")
        return None
````

````python
import pandas as pd
import os

def save_progress(toxic_comments, non_toxic_comments, output_file):

    df = pd.DataFrame({
        'toxic_comment': toxic_comments,
        'non_toxic_comment': non_toxic_comments
    })

    if os.path.exists(output_file):
        df.to_csv(output_file, mode='a', header=False, index=False)
    else:
        df.to_csv(output_file, index=False)
````

````python
import pandas as pd
import os

def save_progress(toxic_comments, non_toxic_comments, output_file):
    df = pd.DataFrame({
        'toxic_comment': toxic_comments,
        'non_toxic_comment': non_toxic_comments
    })

    if os.path.exists(output_file):
        df.to_csv(output_file, mode='a', header=False, index=False)
    else:
        df.to_csv(output_file, index=False)

csv_file_path = 'dataset/augmented_toxic_no_profane_train_3k.csv'
output_csv_file_path = 'dataset/non_toxic_comments_progress2.csv'

data = pd.read_csv(csv_file_path)

if 'comment_text' not in data.columns:
    raise ValueError("The CSV file does not contain a 'comment_text' column.")

toxic_comments = data['comment_text']

if os.path.exists(output_csv_file_path):
    processed_df = pd.read_csv(output_csv_file_path)
    processed_toxic = set(processed_df['toxic_comment'])
else:
    processed_toxic = set()

remaining_toxic = [comment for comment in toxic_comments if comment not in processed_toxic]

batch_size = 50
batch_toxic = []
batch_non_toxic = []
for idx, toxic in enumerate(remaining_toxic, start=1):
    non_toxic = generate_non_toxic(toxic)
    if non_toxic:
        batch_toxic.append(toxic)
        batch_non_toxic.append(non_toxic)

    if idx % batch_size == 0 or idx == len(remaining_toxic):
        save_progress(batch_toxic, batch_non_toxic, output_csv_file_path)
        print(f"Saved {len(batch_toxic)} samples to {output_csv_file_path}")
        batch_toxic = []
        batch_non_toxic = []

print("Processing completed.")
````
